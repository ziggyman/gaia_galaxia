```python
import galcomp
import hammer
import csvData
import matplotlib.pyplot as plt
%matplotlib inline

"""create Hammer object"""
ham = hammer.Hammer()

"""get all pixels on the sky"""
pixels = ham.getPixels()
print pixels[0]
print len(pixels)

"""create rectangular window in xy from which we want the stars"""
pix = hammer.Pixel()
pix.xLow = 2.811
pix.xHigh = 2.825
pix.yLow = 0.1415
pix.yHigh = 0.155
print pix

"""get stars inside our xy window"""
stars = galcomp.getStarsInXYWindow(pixels, pix, "gaia")
print stars.header

"""plot Galactic Latitude vs Galactic Longitude"""
x = stars.getData('l')
y = stars.getData('b')

plt.scatter(x,y)
```
